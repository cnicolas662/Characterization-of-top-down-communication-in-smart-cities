```python
import pandas as pd

Boston = pd.read_excel("DataCoherence.xlsx", sheet_name="Boston")
Helsinki = pd.read_excel("DataCoherence.xlsx", sheet_name="Helsinki")
Seoul = pd.read_excel("DataCoherence.xlsx", sheet_name="Seoul")
Taipei = pd.read_excel("DataCoherence.xlsx", sheet_name="Taipei")

Seoul1 = pd.read_excel("DataCoherence.xlsx", sheet_name="Seoul1")

Taipei1 = pd.read_excel("DataCoherence.xlsx", sheet_name="Taipei1")




```

```python
import numpy as np; np.random.seed(10)
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt


mask1 = np.zeros_like(Taipei)
mask1[np.triu_indices_from(mask1)] = True

mask2 = np.zeros_like(Taipei1)
mask2[np.triu_indices_from(mask2)] = True

mask3 = np.zeros_like(Seoul)
mask3[np.triu_indices_from(mask3)] = True

mask4 = np.zeros_like(Seoul1)
mask4[np.triu_indices_from(mask4)] = True

sns.set(style="white")

cmap = sns.diverging_palette(220, 10, n=30)


f, axes = plt.subplots(2,2, figsize=(11,13))

f.subplots_adjust(top=0.9)

sns.heatmap(Taipei, mask=mask1, cmap = cmap, annot=True, vmax=.2,cbar=False,linewidths=.1,fmt='.2f',   
            square=True, cbar_kws={"shrink": 0.8}, ax =axes[0,0])

sns.heatmap(Taipei1, mask=mask2, cmap = cmap, annot=True, vmax=.4, cbar=False,linewidths=.2,
            square=True, cbar_kws={"shrink": 0.8}, ax =axes[0,1])


sns.heatmap(Seoul, mask=mask3, cmap = cmap, annot=True, vmax=.2, cbar=False,linewidths=.2,
            square=True, cbar_kws={"shrink": 0.8}, ax =axes[1,0])

sns.heatmap(Seoul1, mask=mask4,cmap = cmap, annot=True, vmax=.4, vmin = .2, cbar=False,
            square=True, cbar_kws={"shrink": 0.8}, ax =axes[1,1])



axes[0,0].set_title('Case Study: Taipei',fontdict={'fontsize': 20, 'fontweight': 'bold'})

axes[0,1].set_title('Case Study: Taipei_100',fontdict={'fontsize': 20, 'fontweight': 'bold'})

axes[1,0].set_title('Case Study: Seoul',fontdict={'fontsize': 20, 'fontweight': 'bold'})

axes[1,1].set_title('Case Study: Seoul_100',fontdict={'fontsize': 20, 'fontweight': 'bold'})


plt.setp(axes[0,0].get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.setp(axes[0,1].get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.setp(axes[1,0].get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.setp(axes[1,1].get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")



plt.savefig('Heatmap1_subplots.png')

```
