```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#f,(ax1,ax2) = plt.subplots(2,1, figsize=(15,12))
f,(ax1,ax2) = plt.subplots(1,2, figsize=(14,6))


f.suptitle("Grid-based Coherence Metrics \n", fontsize=20, fontweight='bold')

sns.lineplot(x="Topic", y="Coherence", hue="City",style="City",markers = True,data=df, ax=ax1 )
sns.lineplot(x="Topic", y="Coherence", hue="City",style="City",markers = True,data=df1, ax=ax2 )

#sns.pointplot(x="Topic", y="Coherence",hue="City",  data=df1, ax =ax2)


ax1.set_xlabel("\nNumber of Topics",fontsize=15,fontweight='bold')
ax1.set_ylabel("Coherence C_v",fontsize=15,fontweight='bold')
ax1.tick_params(labelsize=12)

handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles=handles[1:], labels=labels[1:], loc='best',fontsize=15,frameon=False)

ax2.set_xlabel("\nNumber of Topics",fontsize=15,fontweight='bold')
ax2.set_ylabel("Coherence C_UMass",fontsize=15,fontweight='bold')
ax2.tick_params(labelsize=12)
ax2.legend(loc='best', fontsize=12,frameon=False)
handles, labels = ax2.get_legend_handles_labels()
ax2.legend(handles=handles[1:], labels=labels[1:], loc='best',fontsize=15,frameon=False)


sns.set_style("ticks",{"xtick.major.size":8, "ytick.major.size":8})
#sns.set_style("white")
sns.despine()

plt.savefig('Topic_coherence_plot.png')
```
